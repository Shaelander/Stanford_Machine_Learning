# Loc and iloc -- Lecture 6

```python
df.head()
```

```python
#here 0, 1 ,4  are index no of rows
df.iloc[[0,1,4]]
```

```python
#[1:5] is row and 1 is column no
df.iloc[2:4,1]
```

```python
#printing rows with boolean True will be printed and false will be excluded
df.head()
```

```python
df.iloc[[True,False,True,False,True]]
```

```python
# loc gets values on value basis not on index basis like iloc
df.loc[[61,22,60]]
```

```python
#getting values from one colum till another column
df.loc[[61,75],'Name':'Random']
```

```python
df.loc[[61,75],:]
```

```python
# you can do same with loc as well
df.loc[[True,False,True,False,True]]
```

```python
# appling condition like i want values where my random column values are 13>
df.loc[df.Random>13]
```

```python
#using if another condition if the value is greater than 13 or is equal to 10 return values
df.loc[df.Random>13 | (df.Random == 10),:]
```

# Merging part2 Lecture 10

```python
d1 = pd.DataFrame({
    
    "city": ["lakhow","shimla","chandighar","agra","masuri"],
    "temprature":[32,12,20,31,13],
    
})
```

```python
d1
```

```python
d2 = pd.DataFrame({
    
    "city": ["masuri","shimla","chandighar",],
    "Humidity":[55,666,665],
    
})
```

```python
d2
```

```python
df_merge = pd.merge(d1,d2)
```

```python
df_merge
```

```python
#you can also define the kind of join you want to perform
pd.merge(d1,d2 ,on=['city'],how='outer')
```

```python
pd.merge(d1,d2 ,on=['city'],how='right')
```
